```python
# Comparar los valores de las columnas y contar los iguales
cantidad_iguales1 = (df_combinado4['Total'] == df_combinado4['Total_x']).sum()
cantidad_iguales2 = (df_combinado4['Total'] == df_combinado4['Total_y']).sum()
cantidad_iguales3 = (df_combinado4['Total_x'] == df_combinado4['Total_y']).sum()

# Imprimir la cantidad de valores iguales
print(cantidad_iguales1)
print(cantidad_iguales2)
print(cantidad_iguales3) # en este caso coinciden todos los valores
```

Podemos notar que hay diferencia entre los resultados "Totales" procedentes de los distintos dataset originales
En este caso, "Total_x" = "Total_y" (que suman los diferentes tipos de conexiones) pero diferen respecto a "Total" (procedente de la suma de Mbps)
Como coinciden "Total_x" con "Total_y", se elimina la ultima columna.

```python
df_combinado4 = df_combinado4.drop('Total_y', axis=1)
df_combinado4.info()
```

```python
# Renombramos las columnas necesarias
df_combinado4 = df_combinado4.rename(columns={'Total': 'Total suma Mbps'})
df_combinado4 = df_combinado4.rename(columns={'Total_x': 'Total_conexion'})
df_combinado4 = df_combinado4.rename(columns={'OTROS': 'Otros Mbps'})
df_combinado4 = df_combinado4.rename(columns={'HASTA 512 kbps': 'Hasta 512 kbps'})
```

```python
# Revisamos el nuevo df obtenido
df_combinado4.info()
```

```python
# Agrupar y resumir los datos:
df_agrupado = df_combinado4.groupby(['Año', 'Trimestre', 'Provincia']).sum()
df_agrupado.sample(10)
```

# Transformación del segundo df:

- df_nac4

```python
df_nac4.info()
```

Se eliminan columnas redundantes

```python
# Se eliminan las columnas
df_nac5 = df_nac4.copy()
df_nac5 = df_nac5.drop(['Periodo_x', 'Trimestre.1', 'Periodo_y', 'Periodo'], axis=1)
df_nac5.info()
```

```python
# Se revisa la columna "Unnamed: 11" y se constata que la misma está vacia
valores_unicos = df_nac5['Unnamed: 11'].unique()
print (valores_unicos)
```

```python
df_nac5 = df_nac5.drop(['Unnamed: 11'], axis=1)
df_nac5.info()
```

```python
df_nac5.head(10)
```

```python
# Se revisan los valores que coindiden:
cantidad_iguales4 = (df_nac5['Total_x'] == df_nac5['Total_y']).sum()
# Imprimir la cantidad de valores iguales
print(cantidad_iguales4)
```

Existen valores con "." en algunas columnas, se procede a retirarlos

```python
df_nac6 = df_nac5.copy()
df_nac6['Hasta 512 kbps'] = df_nac6['Hasta 512 kbps'].astype(str).str.replace('.', '')
df_nac6.head(5)
```

```python
# Eliminamos del mismo modo "." del resto de columnas
columnas = ['Entre 512 Kbps y 1 Mbps', 'Entre 1 Mbps y 6 Mbps', 'Entre 6 Mbps y 10 Mbps', 'Entre 10 Mbps y 20 Mbps', 'Entre 20 Mbps y 30 Mbps', 'Más de 30 Mbps', 'OTROS']

for columna in columnas:
    df_nac6[columna] = df_nac6[columna].astype(str).str.replace('.', '')
```

```python
# Revisamos las similitudes de las columnas
cantidad_iguales5 = (df_nac6['Total'] == df_nac6['Total_x']).sum()
# Imprimir la cantidad de valores iguales
print(cantidad_iguales5)
```

```python
# Eliminamos columnas con información redundante
df_nac6 = df_nac6.drop(['Total_y', 'Total_x'], axis=1)
df_nac6.head(3)
```

```python
# Renombramos las columnas
df_nac6 = df_nac6.rename(columns={'OTROS': 'Otros Mbps'})
df_nac6 = df_nac6.rename(columns={'Otros': 'Otras conexiones'})
df_nac6 = df_nac6.rename(columns={'Total': 'Total Mbps-conexiones'})
```

```python
# Revisamos el df
df_nac6.info()
```

```python
# se convierten columnas en float
df_nac7 = df_nac6.copy()
columnas_convertir = ['Hasta 512 kbps', 'Entre 512 Kbps y 1 Mbps', 'Entre 1 Mbps y 6 Mbps', 'Entre 6 Mbps y 10 Mbps', 'Entre 10 Mbps y 20 Mbps', 'Entre 20 Mbps y 30 Mbps', 'Más de 30 Mbps', 'Otros Mbps']
df_nac7[columnas_convertir] = df_nac7[columnas_convertir].astype(float)
```

```python
df_nac7.info()
```

```python
df_partidos3 = df_partidos2.drop(['Link Indec','ADSL_y', 'CABLEMODEM_y', 'DIALUP_x','FIBRAOPTICA_x','SATELITAL_y',
                                 'WIRELESS_y','3G_x','4G_x','ADSL','CABLEMODEM','DIALUP_y','FIBRAOPTICA_y', 
                                 '4G_y','3G_y','TELEFONIAFIJA_y','link','WIRELESS','SATELITAL'], axis=1)
df_partidos3.info()
```

```python
df_partidos3['Latitud'] = df_partidos3['Latitud'].str.replace(',', '.')
df_partidos3['Longitud'] = df_partidos3['Longitud'].str.replace(',', '.')
```
